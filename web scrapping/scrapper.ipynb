```python
# Read me -->

# Here i have used the Edge browser driver
# This will scrap the data and out put a file
# for india_export/ india_import data please insert the link here->
# driver.get("-------------link here--------------------")
# My edge browser driver is in my local system, and i have given its path accordingly, please use your driver path


from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait, Select
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
import numpy as np
import pandas as pd
import csv
```

```python
driver= webdriver.Edge("C:/Users/Anand/Documents/internship/task1_scrapper/msedgedriver.exe")
driver.get("https://tradestat.commerce.gov.in/eidb/ecntq.asp")
```

```python


list=[
    '2007-2008',
    '2008-2009',
    '2009-2010',
    '2010-2011',
    '2011-2012',
    '2012-2013',
    '2013-2014',
    '2014-2015',
    '2015-2016',
    '2016-2017',
    '2017-2018',

]
Year = []
Currency =[]
Value=[]

for i in list:
    dropdown= driver.find_element_by_id("select2")
    select1= Select(dropdown)
    try:
        select1.select_by_visible_text(i)
    except NoSuchElementException:
        print("item does not exist")
    
    


    #----------------for getting export in INR---------------
    value = driver.find_element_by_id("radioval")
    value.click()


    submit= driver.find_element_by_id("button1")
    submit.click()

    # num_rows = len (driver.find_elements_by_xpath("/html/body/div/div[2]/div/table/tbody/tr"))
    # num_cols = len (driver.find_elements_by_xpath("/html/body/div/div[2]/div/table/tbody/tr[2]/td"))

    driver.implicitly_wait(5)
    export= driver.find_element_by_xpath("/html/body/div/div[2]/div/table/tbody/tr[last()]/td[5]")
    driver.implicitly_wait(5)

    print(export.text)

    Year.append(i)
    Currency.append("Rs.")
    Value.append(export.text)
    
    driver.implicitly_wait(10)
    # going back to the previous page
    driver.back()
    #================================================================================

    dropdown= driver.find_element_by_id("select2")
    select1= Select(dropdown)
    try:
        select1.select_by_visible_text(i)
    except NoSuchElementException:
        print("item does not exist")
    value= driver.find_element_by_id("radiousd")
    value.click()

    submit= driver.find_element_by_id("button1")
    submit.click()

    export= driver.find_element_by_xpath("/html/body/div/div[2]/div/table/tbody/tr[last()]/td[5]")

    #----------------for getting export in USD---------------

    Year.append(i)
    Currency.append("US$")
    Value.append(export.text)
    
    print(Year)
    print(Currency)
    print(export.text)


    driver.implicitly_wait(10)
    # going back to the previous page
    driver.back()

df= pd.DataFrame({
    "Year" : Year,
    "Currency": Currency,
    "Value" : Value
})
df.to_csv("india_exports.csv", index= False)
```
