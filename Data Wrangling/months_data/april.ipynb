```python
import pandas as pd
```

```python
df_dict = pd.read_excel('data_wrangling_task.xlsx', header=[3,4,5,6] , sheet_name= ['April 2022'])

df = pd.concat(df_dict.values(), axis=0)
df.head()
card_network_operated= df["Card Network Operated"]
```

```python
date=df["Unnamed: 0_level_0"]
date.columns=['Date']
date = date.drop(date.index[25:37])
date['Date'] = pd.to_datetime(date['Date'])
date['Date'] = date['Date'].dt.strftime('%d-%m-%Y')
date
```

```python
credit_card= card_network_operated["Credit Card"]
credit_card=credit_card.dropna()
credit_card = credit_card.drop(credit_card.index[25:37])
credit_card
#credit card transactions at Pos============================================
#--------------------volume--------------------------------
at_PoS=credit_card['at PoS']
pos_vol=at_PoS['Vol']

credit_data_vol= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Credit Card","Channel":"at PoS","Indicator":"Volume"}',
    'Value' :  pos_vol,
    'Units' : 'mn. transactions'
})
credit_data_vol = pd.concat([ date, credit_data_vol],axis=1)
# #--------------------value---------------------------------
pos_val = at_PoS['Val']

credit_data_val= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Credit Card","Channel":"at PoS","Indicator":"Value"}',
    'Value' :  pos_val,
    'Units' : "Rs. '000"
})
credit_data_val = pd.concat([ date, credit_data_val],axis=1)

# #-------------------adding both in one dataframe--------------
credit_data_pos= pd.concat([credit_data_vol,credit_data_val],axis=0)

#============================================================================

# #credit card transaction at e-commerce======================================
at_ecommerce=credit_card['at e-Commerce']
ecommerce_vol=at_ecommerce['Vol']

credit_data_ecommerce_vol= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Credit Card","Channel":"at e-Commerce","Indicator":"Volume"}',
    'Value' :  ecommerce_vol,
    'Units' : 'mn. transactions'
})
credit_data_ecommerce_vol = pd.concat([ date, credit_data_ecommerce_vol],axis=1)
#--------------------value---------------------------------
ecommerce_val = at_ecommerce['Val']

credit_data_ecommerce_val= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Credit Card","Channel":"at e-Commerce","Indicator":"Value"}',
    'Value' :  ecommerce_val,
    'Units' : "Rs. '000"
})
credit_data_ecommerce_val = pd.concat([date, credit_data_ecommerce_val],axis=1)
credit_data_ecommerce_val
#-------------------adding both in one dataframe--------------
credit_data_ecommerce= pd.concat([credit_data_ecommerce_vol,credit_data_ecommerce_val],axis=0)
credit_data_ecommerce

# #=================================ADDING POS AND ECOMMERCE INTO ONE DATAFRAME DEBIT CARD=========================================
credit_card_data= pd.concat([credit_data_pos, credit_data_ecommerce],axis=0)
credit_card_data
```

```python
debit_card=card_network_operated["Debit Card"]
debit_card=debit_card.dropna()
debit_card = debit_card.drop(debit_card.index[25:37])
debit_card
# #credit card transactions at Pos============================================
# #--------------------volume--------------------------------
at_PoS_d=credit_card['at PoS']
pos_vol_d=at_PoS_d['Vol']

debit_data_vol= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Debit Card","Channel":"at PoS","Indicator":"Volume"}',
    'Value' :  pos_vol_d,
    'Units' : 'mn. transactions'
})
debit_data_vol = pd.concat([ date, debit_data_vol],axis=1)
# # #--------------------value---------------------------------
pos_val_d = at_PoS_d['Val']

debit_data_val= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Debit Card","Channel":"at PoS","Indicator":"Value"}',
    'Value' :  pos_val_d,
    'Units' : "Rs. '000"
})
debit_data_val = pd.concat([ date, debit_data_val],axis=1)

# # #-------------------adding both in one dataframe--------------
debit_data_pos= pd.concat([debit_data_vol,debit_data_val],axis=0)
debit_data_pos
# #============================================================================

# # #credit card transaction at e-commerce======================================
at_ecommerce_d=debit_card['at e-Commerce']
ecommerce_vol_d=at_ecommerce['Vol']

debit_data_ecommerce_vol= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Debit Card","Channel":"at e-Commerce","Indicator":"Volume"}',
    'Value' :  ecommerce_vol_d,
    'Units' : 'mn. transactions'
})
debit_data_ecommerce_vol = pd.concat([ date, debit_data_ecommerce_vol],axis=1)
# #--------------------value---------------------------------
ecommerce_val_d = at_ecommerce['Val']

debit_data_ecommerce_val= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Debit Card","Channel":"at e-Commerce","Indicator":"Value"}',
    'Value' :  ecommerce_val_d,
    'Units' : "Rs. '000"
})
debit_data_ecommerce_val = pd.concat([ date, debit_data_ecommerce_val],axis=1)

# #-------------------adding both in one dataframe--------------
debit_data_ecommerce= pd.concat([debit_data_ecommerce_vol,debit_data_ecommerce_val],axis=0)


# #=================================ADDING POS AND ECOMMERCE INTO ONE DATAFRAME DEBIT CARD=========================================
debit_card_data= pd.concat([debit_data_pos, debit_data_ecommerce],axis=0)

debit_card_data
```

```python
PPIs=card_network_operated["Prepaid Payment Instruments (PPIs) Card"]
PPIs.dropna()
PPIs = PPIs.drop(PPIs.index[25:37])
PPIs
# #credit card transactions at Pos============================================
# #--------------------volume--------------------------------
at_PoS_p= PPIs['at PoS']
pos_vol_p=at_PoS_p['Vol']

PPIs_vol= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Prepaid Payment Instruments (PPIs) Card","Channel":"at PoS","Indicator":"Volume"}',
    'Value' :  pos_vol_p,
    'Units' : 'mn. transactions'
})
PPIs_vol = pd.concat([ date, PPIs_vol],axis=1)
# # #--------------------value---------------------------------
pos_val_p = at_PoS_p['Val']

PPIs_val= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Prepaid Payment Instruments (PPIs) Card","Channel":"at PoS","Indicator":"Value"}',
    'Value' :  pos_val_p,
    'Units' : "Rs. '000"
})
PPIs_val = pd.concat([ date, PPIs_val],axis=1)

# # # #-------------------adding both in one dataframe--------------
PPIs_pos= pd.concat([PPIs_vol,PPIs_val],axis=0)
PPIs_pos
# # #============================================================================

# # # #credit card transaction at e-commerce======================================
at_ecommerce_p=PPIs['at e-Commerce']
ecommerce_vol_p=at_ecommerce_p['Vol']

PPIs_ecommerce_vol= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Prepaid Payment Instruments (PPIs) Card","Channel":"at e-Commerce","Indicator":"Volume"}',
    'Value' :  ecommerce_vol_p,
    'Units' : 'mn. transactions'
})
PPIs_ecommerce_vol = pd.concat([ date, PPIs_ecommerce_vol],axis=1)
# # #--------------------value---------------------------------
ecommerce_val_p = at_ecommerce_p['Val']

PPIs_ecommerce_val= pd.DataFrame({
    'Facets' : '{"Category":"Card Network Operated","Type of Card":"Prepaid Payment Instruments (PPIs) Card","Channel":"at e-Commerce","Indicator":"Value"}',
    'Value' :  ecommerce_val_p,
    'Units' : "Rs. '000"
})
PPIs_ecommerce_val = pd.concat([ date, PPIs_ecommerce_val],axis=1)

# #-------------------adding both in one dataframe--------------
PPIs_ecommerce= pd.concat([PPIs_ecommerce_vol,PPIs_ecommerce_val],axis=0)


# # #=================================ADDING POS AND ECOMMERCE INTO ONE DATAFRAME DEBIT CARD=========================================
PPIs_data= pd.concat([PPIs_pos, PPIs_ecommerce],axis=0)

PPIs_data
```

```python
april_data= pd.concat([credit_card_data, debit_card_data, PPIs_data], axis=0)
april_data
april_data.to_csv("april.csv", index=False)
```
